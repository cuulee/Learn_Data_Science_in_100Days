```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('../Data/Mall_Customers.csv')
data.head()
```

```python
#Dropping the column except Income and Spending Score
subset = data.drop(['CustomerID', 'Gender', 'Age'], axis = 1)
```

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")  
dendrogram = sch.dendrogram(sch.linkage(subset, method='ward'))
```

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")  
dendrogram = sch.dendrogram(sch.linkage(subset, method='ward'))
plt.axhline(y=300, color='r', linestyle='--')
```

```python
h_cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
h_cluster.fit_predict(subset)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(subset['Annual Income (k$)'], subset['Spending Score (1-100)'], c=h_cluster.labels_) 
```
